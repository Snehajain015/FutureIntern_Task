```python
#import data
import pandas as pd
#load csv
df=pd.read_csv(r'C:\\Users\\sneha\\Downloads\\Iris.csv')

print(df.head())
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
plt.hist(df['SepalLengthCm'],bins=10,edgecolor='k',alpha=0.7)
plt.xlabel('Sepal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal length')
plt.show()
```

```python
plt.hist(df['SepalWidthCm'],bins=10,edgecolor='k',alpha=0.7)
plt.xlabel('Sepal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python
plt.hist(df['PetalLengthCm'],bins=10,edgecolor='k',alpha=0.7)
plt.xlabel('Petal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal length')
plt.show()
```

```python
plt.hist(df['PetalWidthCm'],bins=10,edgecolor='k',alpha=0.7)
plt.xlabel('Petal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.show()
```

```python
species_count=df['Species'].value_counts()
plt.bar(species_count.index,species_count.values)
plt.xlabel('species')
plt.ylabel('frequency')
plt.title('distribution of species')
plt.show()
```
