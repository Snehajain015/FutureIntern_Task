```python
#import data
import pandas as pd

# Load CSV file
df = pd.read_csv(r'C:\\Users\\sneha\\Downloads\\test.csv')

# Display the first few rows of the dataframe
print(df.head())
```

```python
#identify null values
df.isnull().sum()
```

```python
#deal with the null values     
df['Age'] = df['Age'].fillna(df['Age'].median())
df.dropna(subset=['Fare'],inplace=True)    
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

#identifying outliers using box plot
df['PassengerId'].value_counts().plot(kind='box')
plt.show()

df['Name'].value_counts().plot(kind='box')
plt.show()

df['Sex'].value_counts().plot(kind='box')
plt.show()

df['Parch'].value_counts().plot(kind='box')
plt.show()

df['SibSp'].value_counts().plot(kind='box')
plt.show()

df['Ticket'].value_counts().plot(kind='box')
plt.show()

df['Fare'].value_counts().plot(kind='box')
plt.show()

df['Embarked'].value_counts().plot(kind='box')
plt.show()
```

```python
## now we have to remove this outliers
##i am using IQR method
```

```python
#Removing outliers using IQR method
Q1=df['Fare'].quantile(0.25)
Q3=df['Fare'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df[(df['Fare']>=lower_bound) & (df['Fare']<=upper_bound)]
```
